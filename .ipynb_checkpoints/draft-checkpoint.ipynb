# Data Cleaning

We performed the following steps to prepare our data for analysis:

1. Select relevant columns
2. Convert dataframe column types
3. Handle missing values

```python
import pandas as pd
```

```python
#load the data
df = pd.read_csv('mp.csv', low_memory=False)
df
```

Selecting relevant columns

```python
df = df[["countryname", "edate", "eumember", "pervote", "per108", "per109", "per110", "per607", "per601_2", "per602_2", "per7062"]]
```

```python
#select countries that are EU members
eu_df = df[df['eumember']==10]
eu_df
```

Convert dataframe column types

```python
eu_df['edate'] = pd.to_datetime(eu_df['edate'])
```

Replace missing values

```python
eu_df.fillna(0,inplace=True)
eu_df
```

# Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
```

Visualizing the percentage of votes gained by each party over time

```python
pervote = eu_df.groupby('edate')['pervote'].mean().to_frame()
pervote
```

```python
# Creating a line graph to see the change in percentage of votes gained by each party over the years. 
plt.figure(figsize=(20, 5))
plt.plot(pervote.index, pervote['pervote'], color='blue')
```
